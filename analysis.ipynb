## 2. Background and Related Work

### 2.1 Plato's Republic and the Three-Class System

Plato's *Republic* (c. 375 BCE) presents Kallipolis as a city-state organized into three classes based on the "allegory of metals":

- **Rulers (Gold)**: Philosopher-kings who govern through wisdom and reason
- **Guardians (Silver)**: Warriors who protect the city with courage and discipline  
- **Producers (Bronze/Iron)**: Farmers, craftsmen, and merchants who sustain the economy

Plato argues that justice exists when each class performs its natural function without interfering with others. The Rulers govern, the Guardians protect, and the Producers produce. Modern political philosophy has critiqued this model as inherently classist, since the "ideal harmony" relies on rigid hierarchy and exclusion from governance for most citizens.

### 2.2 Multi-Agent LLM Simulations

Recent advances in AI have enabled sophisticated simulations of social behavior:

- **Park et al. (2023)** in ["Generative Agents: Interactive Simulacra of Human Behavior"](https://arxiv.org/abs/2304.03442) demonstrated that LLM agents can spontaneously exhibit cooperation, relationship formation, and emergent social norms in simulated environments.

- **Aher et al. (2023)** in ["Using Large Language Models to Simulate Multiple Humans and Replicate Human Subject Studies"](https://arxiv.org/abs/2208.10264) showed that LLM agents can replicate classic psychology experiments, suggesting they encode human-like behavioral patterns.

### 2.3 Bias in Large Language Models

LLMs are known to encode societal biases from their training data:

- **Bolukbasi et al. (2016)** in ["Man is to Computer Programmer as Woman is to Homemaker? Debiasing Word Embeddings"](https://arxiv.org/abs/1607.06520) revealed that word embeddings systematically associate professions with gender stereotypes.

- Subsequent research has documented biases related to race, religion, and socioeconomic status in language models. If LLMs encode human biases, then LLM agent societies may reproduce human-like class prejudices.

### 5.1 Define Data Loading Function

We create a reusable function to load and parse log files from any directory. This function:
- Reads all JSONL files from the specified directory
- Extracts crisis context from the user's initial prompt
- Parses salary allocations from the Ruler's SET_SALARY commands
- Maps each role to its Platonic social class

````python
def load_simulation_data(log_dir):
    """
    Load and parse simulation logs from a directory.
    
    Parameters:
    -----------
    log_dir : str
        Path to directory containing JSONL log files
        
    Returns:
    --------
    df_msgs : DataFrame
        All messages from simulations
    df_salary : DataFrame
        Salary allocations per crisis
    """
    files = glob.glob(os.path.join(log_dir, "*.jsonl"))
    
    # Storage
    all_messages = []
    salary_data = []
    
    # Regex patterns
    salary_pattern = re.compile(r"([A-Za-z]+)\s*=\s*(\d+)")
    user_crisis_pattern = re.compile(r"involving\s+(.*)", re.IGNORECASE)
    
    print(f"Found {len(files)} log files in {log_dir}")
    
    for filepath in files:
        with open(filepath, 'r', encoding='utf-8') as f:
            file_messages = []
            current_crisis = "Unknown"
            
            for line in f:
                try:
                    entry = json.loads(line)
                    
                    # Extract crisis from user prompt (primary method)
                    if entry['speaker'] == "user":
                        match = user_crisis_pattern.search(entry['message'])
                        if match:
                            raw_crisis = match.group(1).strip().strip('."\'')
                            current_crisis = raw_crisis.capitalize()
                    
                    # Fallback: Extract from God's JSON crisis announcement
                    elif entry['speaker'] == "God" and "crisis" in entry['message'] and current_crisis == "Unknown":
                        try:
                            if '{' in entry['message']:
                                json_str = re.search(r"\{.*\}", entry['message']).group(0)
                                msg_json = json.loads(json_str)
                                current_crisis = msg_json.get('crisis', 'Unknown')[:50] + "..."
                        except:
                            pass
                    
                    # Capture salary allocations from Ruler's SET_SALARY command
                    if entry['speaker'] == "Philosopher_Ruler" and "SET_SALARY" in entry['message']:
                        try:
                            raw_text = entry['message'].split("SET_SALARY")[1]
                            matches = salary_pattern.findall(raw_text)
                            for role, amount in matches:
                                salary_data.append({
                                    "File": os.path.basename(filepath),
                                    "Crisis": current_crisis,
                                    "Role": role,
                                    "Salary": int(amount)
                                })
                        except IndexError:
                            pass
                    
                    # Store message with context
                    entry['crisis_context'] = current_crisis
                    entry['filename'] = os.path.basename(filepath)
                    file_messages.append(entry)
                    
                except json.JSONDecodeError:
                    continue
            
            all_messages.extend(file_messages)
    
    # Create DataFrames
    df_msgs = pd.DataFrame(all_messages)
    df_salary = pd.DataFrame(salary_data)
    
    # Define Platonic class mappings
    class_map = {
        "Philosopher_Ruler": "Ruling Class",
        "Warrior": "Guardian Class",
        "Farmer": "Producer Class",
        "Builder": "Producer Class",
        "Merchant": "Producer Class",
        "Artist": "Producer Class",
        "Healer": "Producer Class",
        "Teacher": "Producer Class",
        "God": "System"
    }
    
    # Apply class mappings
    if not df_msgs.empty:
        df_msgs['Social_Class'] = df_msgs['speaker'].map(class_map)
    if not df_salary.empty:
        df_salary['Social_Class'] = df_salary['Role'].map(class_map)
    
    print(f"Loaded {len(df_msgs)} messages and {len(df_salary)} salary records.")
    
    return df_msgs, df_salary
````

### 6.5 Linguistic Marker Analysis

**Research Question:** Do citizens exhibit linguistic patterns of deference toward the Ruler, and does the Ruler use dominance language?

We analyze the frequency of:
- **Deference markers**: Words suggesting submission or respect (e.g., "humble," "honored," "wise")
- **Dominance markers**: Words suggesting authority or command (e.g., "must," "command," "decree")

````python
import nltk
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon', quiet=True)

# Define linguistic marker lists
deference_markers = [
    'majesty', 'lord', 'sire', 'honored', 'wise', 'humble',
    'permission', 'concur', 'agree', 'guidance', 'support', 'appreciate'
]

dominance_markers = [
    'must', 'command', 'order', 'require', 'ensure', 'deploy',
    'secure', 'directive', 'establish', 'prioritize', 'mobilize', 'decree'
]

def count_markers(text, markers):
    """Count occurrences of marker words in text."""
    text = str(text).lower()
    return sum(1 for word in markers if word in text)

# Apply marker analysis to Phase 1 data
df_msgs_1['deference_score'] = df_msgs_1['message'].apply(lambda x: count_markers(x, deference_markers))
df_msgs_1['dominance_score'] = df_msgs_1['message'].apply(lambda x: count_markers(x, dominance_markers))

# Apply sentiment analysis
sia = SentimentIntensityAnalyzer()
df_msgs_1['sentiment_compound'] = df_msgs_1['message'].apply(lambda x: sia.polarity_scores(str(x))['compound'])

# Aggregate by role
linguistic_summary_1 = df_msgs_1.groupby('speaker').agg({
    'deference_score': 'mean',
    'dominance_score': 'mean',
    'sentiment_compound': 'mean'
}).sort_values('dominance_score', ascending=False)

print("--- Phase 1: Linguistic Marker Analysis ---")
display(linguistic_summary_1.round(3))

# --- Visualization Section ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# FILTER: Create a subset of data excluding 'System' for the plots
df_plot = df_msgs_1[df_msgs_1['Social_Class'] != 'System']

# Deference scores (using df_plot instead of df_msgs_1)
deference_by_class = df_plot.groupby('Social_Class')['deference_score'].mean().sort_values(ascending=False)
deference_by_class.plot(kind='bar', ax=axes[0], color=[class_palette.get(x, 'gray') for x in deference_by_class.index])
axes[0].set_title('Phase 1: Deference Language by Class', fontweight='bold')
axes[0].set_ylabel('Average Deference Score')
axes[0].tick_params(axis='x', rotation=45)

# Dominance scores (using df_plot instead of df_msgs_1)
dominance_by_class = df_plot.groupby('Social_Class')['dominance_score'].mean().sort_values(ascending=False)
dominance_by_class.plot(kind='bar', ax=axes[1], color=[class_palette.get(x, 'gray') for x in dominance_by_class.index])
axes[1].set_title('Phase 1: Dominance Language by Class', fontweight='bold')
axes[1].set_ylabel('Average Dominance Score')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('visualizations/phase1_linguistic_markers.png', dpi=150, bbox_inches='tight')
plt.show()
````

### 6.6 Influence Measurement (BERT + ROUGE-L Combined Score)

**Research Question:** Whose advice does the Ruler actually follow?

We measure citizen influence using a **combined metric** that captures both:
- **BERT F1 Score**: Semantic similarity (does the meaning align?)
- **ROUGE-L Score**: Lexical overlap (did the Ruler use similar words/phrases?)

The combined influence score averages both metrics, providing a robust measure of how much each citizen's suggestion shaped the final directive.

````python
import pandas as pd
import matplotlib.pyplot as plt
from rouge_score import rouge_scorer
from bert_score import score as bertscore
import warnings

# Suppress HuggingFace warnings for cleaner output
warnings.filterwarnings("ignore")

rouge_evaluator = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)

def calculate_hybrid_score(citizen_text, ruler_text):
    """
    Helper function to calculate combined ROUGE-L and BERTScore.
    """
    if not citizen_text or not ruler_text:
        return 0.0, 0.0, 0.0

    # 1. ROUGE-L (Literal Overlap)
    rouge_scores = rouge_evaluator.score(ruler_text, citizen_text)
    rouge_l = rouge_scores['rougeL'].fmeasure

    # 2. BERTScore (Semantic Overlap)
    try:
        P, R, F1 = bertscore([citizen_text], [ruler_text], lang="en", verbose=False, model_type='distilbert-base-uncased')
        bert_f1 = F1.mean().item()
    except Exception as e:
        print(f"BERTScore Error: {e}")
        bert_f1 = 0.0

    # 3. Combined Score (Equal Weighting)
    combined = (0.5 * rouge_l) + (0.5 * bert_f1)
    
    return rouge_l, bert_f1, combined

def calculate_influence_scores(df_msgs):
    """
    Calculate combined BERT + ROUGE-L influence scores between citizen suggestions 
    and Ruler directives using the updated hybrid scoring method.
    """
    # Get unique simulation files
    sessions = df_msgs['filename'].unique()
    print(f"Processing {len(sessions)} simulation sessions for Influence Analysis...")
    
    influence_records = []
    
    for session in sessions:
        session_msgs = df_msgs[df_msgs['filename'] == session]
        
        # Get Ruler's directive (message containing 'directive')
        ruler_msgs = session_msgs[session_msgs['speaker'] == 'Philosopher_Ruler']
        directive_msgs = ruler_msgs[ruler_msgs['message'].str.contains('directive', case=False, na=False)]
        
        if directive_msgs.empty:
            continue
            
        directive_text = directive_msgs.iloc[-1]['message']  # Last directive
        crisis = session_msgs.iloc[0]['crisis_context']
        
        # Get citizen messages (exclude Ruler, God, user, SYSTEM)
        citizen_msgs = session_msgs[~session_msgs['speaker'].isin(
            ['Philosopher_Ruler', 'God', 'user', 'SYSTEM']
        )]
        
        if citizen_msgs.empty:
            continue
        
        for _, row in citizen_msgs.iterrows():
            citizen_text = str(row['message'])
            
            # Use the hybrid score function
            rouge_l, bert_f1, combined_influence = calculate_hybrid_score(citizen_text, directive_text)
            
            influence_records.append({
                'File': session,
                'Crisis': crisis,
                'Role': row['speaker'],
                'Social_Class': row['Social_Class'],
                'BERT_F1': bert_f1,
                'ROUGE_L': rouge_l,
                'Combined_Influence': combined_influence
            })
    
    return pd.DataFrame(influence_records)

# Calculate influence
influence_df_1 = calculate_influence_scores(df_msgs_1)

# Display average influence by role (showing all three metrics)
influence_summary_1 = influence_df_1.groupby('Role').agg({
    'BERT_F1': 'mean',
    'ROUGE_L': 'mean',
    'Combined_Influence': 'mean'
}).sort_values('Combined_Influence', ascending=False)

print("\n--- Phase 1: Average Influence Scores by Role ---")
display(influence_summary_1.round(3))

# Visualize combined influence
influence_by_role_1 = influence_df_1.groupby('Role')['Combined_Influence'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
# Get colors based on the social class of the Role
colors = [class_palette.get(
    influence_df_1[influence_df_1['Role'] == role]['Social_Class'].iloc[0], 'gray'
) for role in influence_by_role_1.index]

influence_by_role_1.plot(kind='bar', color=colors)
plt.title('Phase 1: Citizen Influence on Ruler\'s Directive (BERT + ROUGE-L Combined)', fontsize=12, fontweight='bold')
plt.xlabel('Role')
plt.ylabel('Combined Influence Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('visualizations/phase1_influence.png', dpi=150, bbox_inches='tight')
plt.show()
````

````python
from scipy import stats

# Merge influence with salary data
merged_1 = pd.merge(
    influence_df_1.groupby(['File', 'Role'])['Combined_Influence'].mean().reset_index(),
    df_salary_1,
    on=['File', 'Role'],
    how='inner'
)

print(f"Phase 1: Merged {len(merged_1)} data points for correlation analysis.")

# Calculate correlations
pearson_r_1, pearson_p_1 = stats.pearsonr(merged_1['Combined_Influence'], merged_1['Salary'])
spearman_r_1, spearman_p_1 = stats.spearmanr(merged_1['Combined_Influence'], merged_1['Salary'])

print(f"\n--- Phase 1: Influence vs. Salary Correlation ---")
print(f"Pearson Correlation:  r = {pearson_r_1:.4f} (p = {pearson_p_1:.4e})")
print(f"Spearman Correlation: ρ = {spearman_r_1:.4f} (p = {spearman_p_1:.4e})")

# Interpretation
if abs(pearson_r_1) < 0.1:
    interpretation_1 = "NO CORRELATION: Salaries appear disconnected from actual contribution."
elif abs(pearson_r_1) < 0.3:
    interpretation_1 = "WEAK CORRELATION: Some relationship exists, but other factors dominate."
else:
    interpretation_1 = "MODERATE/STRONG CORRELATION: Contribution influences rewards."

print(f"\n*** INTERPRETATION: {interpretation_1} ***")

# Visualize
plt.figure(figsize=(8, 6))
for social_class in merged_1['Social_Class'].unique():
    subset = merged_1[merged_1['Social_Class'] == social_class]
    plt.scatter(
        subset['Combined_Influence'],
        subset['Salary'],
        c=class_palette.get(social_class, 'gray'),
        label=social_class,
        alpha=0.7,
        s=60
    )

# Add regression line
z = np.polyfit(merged_1['Combined_Influence'], merged_1['Salary'], 1)
p = np.poly1d(z)
x_line = np.linspace(merged_1['Combined_Influence'].min(), merged_1['Combined_Influence'].max(), 100)
plt.plot(x_line, p(x_line), "r--", alpha=0.8, label=f'Trend (r={pearson_r_1:.2f})')

plt.title('Phase 1: Influence vs. Salary (Is Reward Based on Contribution?)', fontsize=12, fontweight='bold')
plt.xlabel('Combined Influence Score (BERT + ROUGE-L)')
plt.ylabel('Salary (Gold)')
plt.legend()
plt.tight_layout()
plt.savefig('visualizations/phase1_correlation.png', dpi=150, bbox_inches='tight')
plt.show()
````

## 7. The Discovery: Prompt Contamination

While investigating the Phase 1 results, I discovered a critical flaw in the Ruler's system prompt. The prompt included mathematical instructions for balancing salaries, with a specific example:

```
**ZERO SUM:** If you give the Warrior 140, you MUST give two others 80 to balance it.
```

### The Problem

By explicitly mentioning **"Warrior"** in the example (rather than a generic placeholder like "citizen A"), the prompt inadvertently **primed** the language model to:
1. Associate the Warrior role with the "high salary" example
2. Anchor the Warrior as a reference point for generous compensation
3. Subtly bias salary decisions toward the Guardian class

This is a textbook case of **prompt injection bias** — where a seemingly neutral instruction contains implicit cues that influence model behavior.

### The Fix

The corrected prompt replaced the specific role name with an abstract placeholder:

```
**ZERO SUM:** If you give citizen A a salary of 140, you MUST give two others 80 to balance it.
```

### Hypothesis

With the biased example removed, salary allocations should become:
1. More context-sensitive (relevant roles rewarded for relevant crises)
2. More meritocratic (correlation between influence and salary should increase)
3. Less systematically biased toward the Warrior

The following section analyzes the Phase 2 simulations to test this hypothesis.

## 10. Discussion

### 10.1 Answering the Research Questions

**RQ1 (Linguistic Bias):** Do LLM agents develop class-based linguistic biases?

> **Finding:** Yes, but moderately. The Ruler exhibits significantly higher dominance language scores, while citizens show deference patterns. However, within the citizen class, linguistic differences were subtle. The more striking finding was the economic bias in salary allocation.

**RQ2 (Ruler Favoritism):** Does the Ruler favor certain classes in decision-making?

> **Finding:** In Phase 1, the Warrior received consistently high salaries regardless of crisis context, suggesting favoritism. After correcting the prompt, salary allocation became more context-sensitive, with relevant experts (Farmer for crop crises, Healer for health crises) receiving appropriate recognition.

**RQ3 (Moral Language):** Are moral markers distributed unevenly?

> **Finding:** The Ruler's language contains more directive/authoritative markers, while citizen speech contains more deferential language. This reflects the designed hierarchy, but also suggests LLMs reproduce expected power dynamics in language.

**RQ4 (Meritocracy vs. Bias):** Does influence correlate with reward?

> **Finding:** This was the most revealing metric:
> - **Phase 1:** r = 0.04 (no correlation) — Rewards were disconnected from contribution
> - **Phase 2:** r = 0.40 (moderate, significant correlation) — Removing prompt bias improved meritocracy

### 10.2 The Meta-Finding: Prompt Bias Amplification

Perhaps the most important discovery was not about Plato's Kallipolis, but about **LLM prompt design**. A single, seemingly innocuous example in the system prompt ("If you give the Warrior 140...") created measurable bias in agent behavior across all simulations.

This has significant implications:
1. **LLMs are highly sensitive to examples in prompts**, even when those examples are meant to be generic
2. **Prompt designers must carefully audit examples** for implicit biases
3. **Abstract placeholders ("citizen A") are safer** than specific names in instructional examples

### 10.3 Limitations

This study has several limitations that should be considered:

1. **Model Specificity:** Results are based on `llama3.1:8b-instruct-q8_0`. Different models may exhibit different bias patterns.

2. **Sample Size:** 12 crises × 2 phases = 24 total simulations. Larger samples would increase statistical power.

3. **Crisis Selection:** The 12 crises were author-selected and may not represent all possible scenarios.

4. **Single-Run Variability:** LLM outputs are stochastic. Multiple runs per crisis would better characterize variability.

5. **Prompt Complexity:** The system prompts are complex, and multiple factors could influence behavior.

### 10.4 Implications

**For AI System Design:**
- Multi-agent LLM systems can inadvertently reproduce human-like biases
- Prompt auditing should be a standard practice before deployment
- Even "virtuous" instructions cannot fully override implicit biases in examples

**For Philosophy and Political Theory:**
- The question of whether ideal societies remain just is as relevant for AI systems as for humans
- LLM agents, while not human, encode human patterns and prejudices
- "Artificial" does not mean "unbiased"

**For Data Science Practice:**
- Unexpected confounding variables (like prompt examples) can significantly affect results
- Transparency about methodology changes (like the prompt fix) strengthens research credibility
- Natural experiments (before/after) can emerge from debugging processes

## 11. Conclusion

This study set out to investigate whether LLM agents, when organized into Plato's hierarchical class structure, would exhibit classism or favoritism — even when instructed to act virtuously.

### Summary of Findings

1. **Initial Bias Detected:** In Phase 1, the Warrior (Guardian Class) received a 6% salary premium over Producers, and there was no correlation between citizen influence and economic reward.

2. **Prompt Contamination Discovered:** A specific example in the system prompt ("If you give the Warrior 140...") inadvertently primed the model to favor Warriors in salary allocation.

3. **Correction Improved Meritocracy:** After replacing "Warrior" with the neutral "citizen A," the salary-influence correlation improved from r=0.04 to r=0.40, and the Warrior Premium dropped from 1.06x to 1.01x.

4. **"The Artist Exception"**: Unlike the Warrior bias (which was artificial), the consistently low Artist salary represents appropriate context-sensitivity — art genuinely contributes less to immediate crisis resolution. This demonstrates the model can make rational utility-based distinctions when not contaminated by prompt bias.

5. **Context-Sensitivity Emerged:** Post-correction, salary allocation became more appropriate to crisis context (e.g., Farmers highest paid during crop failures).

### The Bigger Picture

Plato argued that Kallipolis would remain just because its citizens, guided by reason, would fulfill their natural roles without corruption. This study suggests that LLM agents can approximate this ideal — but only when their instructions are carefully purged of implicit biases.

The unexpected lesson is not just about ancient philosophy, but about modern AI: **the biases we encode in prompts, even unintentionally, will manifest in agent behavior**. As LLM-based systems become more prevalent in decision-making contexts, this finding underscores the critical importance of careful prompt design and bias auditing.

### Future Work

Future research could extend this work by:
- Testing across multiple LLM architectures
- Running longer simulations (dozens of "years") to observe drift
- Introducing explicit conflict between classes
- Comparing hierarchical (aristocratic) vs. flat (democratic) structures
- Developing automated prompt bias detection tools
